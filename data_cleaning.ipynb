```python
cleaned_data = csv_data[["age", "sex", "intraop_ebl", "ane_type", "department", "opstart", "opend", "anestart", "aneend", "asa", "approach", "death_inhosp"]]
cleaned_data['surgery_duration'] = (cleaned_data['opend'] - cleaned_data['opstart']) / 60
cleaned_data['anes_duration'] = (cleaned_data['aneend'] - cleaned_data['anestart']) / 60
cleaned_data = cleaned_data.drop(columns=['opstart', 'opend', 'anestart', 'aneend'])
cleaned_data['mortality_label'] = cleaned_data["death_inhosp"].apply(lambda x: "Survived" if x == 0 else "Mortality")
cleaned_data['color'] = cleaned_data["death_inhosp"].apply(lambda x: "green" if x == 0 else "red")
cleaned_data
```

```python
df['adm'] = df['adm'].apply(lambda x: abs(x) if x < 0 else x)  # Fix negative admission times
df['dis'] = df['dis'].apply(lambda x: x if x > 0 else None)  # Remove negative discharge times

# Compute los_postop after fixing adm & dis
df['los_postop'] = (df['dis'] - df['adm']) / (60 * 60 * 24)  # Convert to days

# Handle invalid values
df.loc[df['los_postop'] < 0, 'los_postop'] = None  # Remove negatives
df['los_postop'] = df['los_postop'].fillna(df['icu_days'])  # Use ICU stay if missing

# Check final results
print(df[['adm', 'dis', 'los_postop', 'icu_days']].head())
# Ensure mortality data is numeric
df['death_inhosp'] = pd.to_numeric(df['death_inhosp'], errors='coerce')

# Convert to categorical labels
df['mortality_label'] = df['death_inhosp'].map({0: "Survived", 1: "Mortality"})

# Assign colors for visualization
df['color'] = df['death_inhosp'].map({0: "green", 1: "red"})

df
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df = df[df['age'] > 0]  # Remove invalid ages
df['sex'] = df['sex'].str.upper().fillna("UNKNOWN")  # Standardize values
df['bmi'] = pd.to_numeric(df['bmi'], errors='coerce')
df = df[(df['bmi'] >= 10) & (df['bmi'] <= 60)]
df['intraop_ebl'] = pd.to_numeric(df['intraop_ebl'], errors='coerce')
df = df.assign(intraop_ebl=df['intraop_ebl'].fillna(df['intraop_ebl'].median()))
df = df[df['intraop_ebl'] >= 0]
df['surgery_duration'] = (df['opend'] - df['opstart']) / 60  # Convert sec â†’ min
df = df[df['surgery_duration'] >= 0]  # Remove invalid cases
df['asa'] = pd.to_numeric(df['asa'], errors='coerce')
df = df[df['asa'].between(1, 5, inclusive='both')]  # ASA is between 1 and 5
df['approach'] = df['approach'].fillna("Unknown")
columns_to_keep = ['age', 'sex', 'bmi', 'department', 'intraop_ebl', 'surgery_duration', 'anes_duration', 'ane_type',
                   'asa', 'approach', 'los_postop', 'death_inhosp', 'mortality_label', 'color']

df_filtered = df[columns_to_keep]
df_filtered.to_csv("clinical_data_cleaned.csv", index=False)
df_filtered.to_json("clinical_data_cleaned.json", orient="records")
```
